```python
import pandas as pd
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt

# Data Preprocessing
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical

# Building CNN Model & Pretraining Model
import tensorflow as tf
import tensorflow.keras as k
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, AveragePooling2D, Flatten, Dropout

# Download Model
import pickle
```

```python
path_folder = "/content/drive/My Drive/Hack-Summit/dataset/New-Plant-Diseases-Dataset-Augmented/whatever/train"
class_name = os.listdir(path_folder)
class_name.sort()
class_name = class_name[:13] + class_name[20:23]
print(class_name)
image_data = []
label_data = []
count = 0
for folder in class_name:
    images = os.listdir(path_folder + "/" + folder)
    print("Loading Folder -- {} " .format(folder), "The Count of Classes ==> ",count)
    for img in images:
        image = cv2.imread(path_folder + "/" + folder + "/" + img)
        image = cv2.resize(image, (100, 100))

        image_data.append(image)
        label_data.append(count)
    count += 1
print("---- Done ----------- ")
```

```python
data = np.array(image_data)
data = data.astype("float32")
data = data/255.0

label = np.array(label_data)
```

```python
print(data.shape)
```

```python
label_num = to_categorical(label, len(class_name))
label_num[100]
```

```python
x_img, y_img = shuffle(data, label_num)
x_train, x_test, y_train, y_test = train_test_split(x_img, y_img, train_size=0.8)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
plt.figure(figsize=(10, 10))
for i in range(0, 25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_train[i])
    plt.title(class_name[np.argmax(y_train[i])])
```

```python
model = k.models.Sequential()

model.add(k.layers.Conv2D(16, (5, 5), activation="relu", input_shape=(100, 100, 3), padding="same"))
model.add(k.layers.AveragePooling2D((2, 2)))

model.add(k.layers.Conv2D(32, (4, 4), activation="relu",  padding="same"))
# model.add(k.layers.BatchNormalization())
model.add(k.layers.AveragePooling2D((2, 2)))


model.add(k.layers.Conv2D(64, (3, 3), activation="relu", padding="same"))
model.add(k.layers.AveragePooling2D((2, 2)))

model.add(k.layers.Conv2D(128, (2, 2), activation="relu", padding="same"))
model.add(k.layers.MaxPool2D((2, 2)))

model.add(k.layers.Flatten())

model.add(k.layers.Dense(256, activation="relu"))
# model.add(k.layers.BatchNormalization())
model.add(k.layers.Dropout(0.5))

model.add(k.layers.Dense(32, activation="relu"))
model.add(k.layers.Dropout(0.2))

model.add(k.layers.Dense(13, activation="softmax"))

model.compile(optimizer="adam", loss=k.losses.CategoricalCrossentropy(), metrics=["accuracy"])

model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=50, validation_data=(x_test, y_test), validation_split=0.4)
```

```python
plt.plot(history.history["loss"], label="Loss")
plt.plot(history.history["val_loss"], label="Val_Loss")

plt.xlabel("Epochs")
plt.ylabel("Loss")

plt.title("Loss Vs Epochs")

plt.legend()
plt.grid()
```

```python
plt.plot(history.history["accuracy"], label="accuracy")
plt.plot(history.history["val_accuracy"], label="val_accuracy")

plt.xlabel("Epochs")
plt.ylabel("Accuracy")

plt.title("Accuracy Vs Epochs")

plt.legend()
plt.grid()
```

```python
loss, accuracy = model.evaluate(x_test, y_test)

print(f"The Accuracy Testing Data is ==> {accuracy} %")
print(f"The loss Testing Data is ==> {loss}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Input the file path to the image on your local machine
file_path = input("Enter the path to the image: ")

# Read and preprocess the image with the correct bit depth (8 bits per channel)
image = cv2.imread(file_path, cv2.IMREAD_COLOR)  # Use cv2.IMREAD_COLOR to ensure 8-bit depth

# Check if the image was successfully loaded
if image is not None:
    # Check the bit depth of the loaded image
    print("Image dtype:", image.dtype)

    # Ensure the image has the correct bit depth (convert to 8 bits per channel)
    if image.dtype != np.uint8:
        image = cv2.convertScaleAbs(image)

    # Resize to match the model's input shape
    image = cv2.resize(image, (100, 100))
    image = image / 255.0  # Normalize pixel values

    # Normalize the image to 0-255 range and convert to uint8
    image = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    # Reshape the image to match the model's input shape
    new_image = np.reshape(image, (1, 100, 100, 3))

    # ================== Show Prediction =================================

    # Make a prediction using the model
    prediction = model.predict(new_image)[0]

    # Get the predicted class index
    predicted_class_index = np.argmax(prediction)

    # Extract the predicted class name from the class_name list
    new_prediction = class_name[predicted_class_index]

    # Split the predicted class name into "Type Of Plant" and "Type Of Disease" components
    split_prediction = new_prediction.split("__")

    # Print the results
    print("Type Of Plant:", split_prediction[0])
    print("Type Of Disease:", split_prediction[1])

    # ================== Show Image And Label Prediction =================================

    # Convert the image to RGB format for displaying with matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display the image with the predicted label as the title
    plt.imshow(image_rgb)
    plt.title(new_prediction)
    plt.show()

else:
    print("Failed to load the image. Please check the file path.")

```
